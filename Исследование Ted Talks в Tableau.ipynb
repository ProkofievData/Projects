# Описание проекта

В этом проекте я исследовал историю TED-конференций с помощью Tableau.

# Проделанные шаги


## Шаг 1. Загрузка и изучение данных

Я загрузил данные в Tableau и с помощью Union объединил файлы `tableau_project_data_1.csv`, `tableau_project_data_2.csv` и `tableau_project_data_3.csv` в единую таблицу. Затем установил связи (Relationships) с таблицами-справочниками: `tableau_project_event_dict.csv` и `tableau_project_speakers_dict.csv`.

Изучил состав и типы данных. С помощью визуализации определил, как менялось количество конференций во времени, и нашёл пороговый год, после которого наблюдается скачкообразный рост числа выступлений. Установил фильтр, исключающий все выступления до этого года на уровне источника данных.

## Шаг 2. Дашборд «История выступлений»

Создал:

- лист **«Выступления по странам»** с картой, отображающей процент выступлений по странам
- лист **«Выступления по годам»** с area chart по числу выступлений в год, разбитых по топ-N тематикам (создал соответствующий сет)
- копию этого листа под названием **«Процент выступлений по тематикам»**, в котором отобразил долю тематик по годам
- фильтр по полю `Film Date`, применил его ко всем визуализациям

Собрал дашборд **«История выступлений»** по предложенному макету. Настроил Action: при выборе страны фильтр применяется ко всем листам.

В блоке выводов указал:

- в каких странах чаще всего проходили выступления
- какие категории были наиболее популярны
- как изменился состав тематик в 2020 и 2021 годах

## Шаг 3. Дашборд «Тематики выступлений»

Создал:

- лист **«Выступления по тематике»** с круговой диаграммой топ-10 категорий + «Другие» (с помощью сета)
- лист **«Тематики и страны»** с highlight table, в строках — топ-10 тематик + «Другие», в столбцах — топ-3 страны (сеты), значения — процент тематики в каждой стране. Отсортировал таблицу по числу просмотров
- вычисляемые поля:
  - `Duration, min` — длительность в минутах
  - `Applause by Duration` — аплодисменты на секунду
  - `Laughter by Duration` — смех на секунду
- диаграммы размаха:
  - **«Аплодисменты по тематикам»** — Applause by Duration, отсортировано по медиане
  - **«Смех по тематикам»** — аналогично, но по Laughter by Duration
- гистограммы:
  - **«Гистограмма длительности»**
  - **«Гистограмма просмотров»** — обе показывают проценты наблюдений
- добавил гистограммы во всплывающую подсказку на листе «Выступления по тематике»
- **«Связь длительности с просмотрами»** — scatter plot с длительностью по оси X и просмотрами по оси Y, оформил всплывающую подсказку

Создал дашборд **«Тематики выступлений»** по макету. Настроил Action: выбор тематики фильтрует все визуализации, кроме таблицы «Тематики и страны».

В области выводов ответил:

- какие тематики наиболее популярны
- различается ли распределение популярных категорий в разных странах
- какие темы вызывают больше/меньше аплодисментов и смеха
- какое выступление оказалось самым смешным, а какое — с наибольшими аплодисментами
- есть ли связь между длительностью и просмотрами
- какое выступление самое продолжительное и самое просматриваемое

## Шаг 4. Дашборд «Авторы выступлений»

Создал:

- параметр `Top-N Speaker Occupations` с элементом управления Slider
- сет топ-N популярных профессий
- лист **«Области деятельности авторов»** с пузырьковой диаграммой (без категории «Другие»)
- вычисляемое поле `Talks by Author` (LOD)
- **«Распределение числа выступлений»** — столбчатая диаграмма
- **«Авторы по числу выступлений»** — таблица с именем, профессией, описанием, числом выступлений, отсортированная по убыванию
- **«Выступления и число просмотров»** — таблица с названием, описанием, конференцией и числом просмотров, отсортирована по убыванию

Создал дашборд **«Авторы выступлений»** по макету. Настроил Actions:

- при выборе профессии фильтруются обе таблицы
- при выборе автора фильтруется таблица выступлений

Добавил элемент управления параметром `Top-N Speaker Occupations`.

В блоке выводов ответил:

- какие профессии преобладают
- сколько выступлений приходится на автора в среднем
- кто выступал чаще всех
- какой дизайнер был самым активным и какое его выступление смотрели меньше всего

## Шаг 5. Дашборд на свободную тему

Добавил дашборд на свободную тему с двумя визуализациями. В моем случае — анализ состава категории «Другие» в 2019–2021 годах. Добавил выводы и наблюдения.
